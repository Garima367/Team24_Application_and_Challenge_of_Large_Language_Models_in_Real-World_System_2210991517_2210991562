```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyArrowPatch, Circle

fig, ax = plt.subplots(figsize=(12,5))

# ---------------------------
# Helper functions
# ---------------------------
def draw_text(x, y, text, size=12, weight='normal'):
    ax.text(x, y, text, ha='center', va='center',
            fontsize=size, fontweight=weight)

def draw_arrow(x1, y1, x2, y2):
    ax.add_patch(FancyArrowPatch((x1,y1),(x2,y2),
                                arrowstyle='->', mutation_scale=15))

def dot(x, y):
    c = Circle((x,y), 0.06)
    ax.add_patch(c)

# ---------------------------
# Title
# ---------------------------
draw_text(5, 3.8, "FINETUNING PROCESS", size=24, weight='bold')

# ---------------------------
# Main pipeline (horizontal)
# ---------------------------
y = 2

# Raw data
draw_text(0.8, y-0.6, "Raw Text Data")
dot(1.5, y)

# Pre-training
draw_text(2.5, y, "Pre - training")
draw_arrow(1.6, y, 3.5, y)

# Pre-trained model
draw_text(4.5, y-0.6, "Pre - Trained LLM")

# Fine-tuning step
draw_arrow(5.5, y, 7, y)
draw_text(6.2, y, "Fine - Tuning")

# Final model
draw_arrow(7.5, y, 9, y)
draw_text(9.5, y-0.6, "Fine - Tuned LLM")

# ---------------------------
# Custom Knowledge (vertical)
# ---------------------------
draw_text(6.2, 3.0, "Custom\nKnowledge Base")
dot(6.2, 2.6)
draw_arrow(6.2, 2.6, 6.2, 2.2)

# ---------------------------
# Styling
# ---------------------------
ax.set_xlim(0, 10.5)
ax.set_ylim(1, 4.2)
ax.axis('off')

plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch, Rectangle, FancyArrowPatch

fig, ax = plt.subplots(figsize=(12,6))

# -------------------------
# Helper functions
# -------------------------
def draw_box(x, y, w, h, text, fc='white'):
    rect = Rectangle((x, y), w, h, linewidth=1, edgecolor='black', facecolor=fc)
    ax.add_patch(rect)
    ax.text(x + w/2, y + h/2, text, ha='center', va='center', fontsize=10)

def draw_round_box(x, y, w, h, text, fc):
    rect = FancyBboxPatch((x, y), w, h,
                          boxstyle="round,pad=0.3",
                          linewidth=1, edgecolor='black', facecolor=fc)
    ax.add_patch(rect)
    ax.text(x + w/2, y + h/2, text, ha='center', va='center', fontsize=11, color='white')

def arrow(x1, y1, x2, y2):
    ax.add_patch(FancyArrowPatch((x1,y1),(x2,y2),
                                arrowstyle='->', mutation_scale=15))

# -------------------------
# Title
# -------------------------
ax.text(5, 5.5, "Tokenizer Pipeline", ha='center', fontsize=18, color='red')

# -------------------------
# Outer big box
# -------------------------
outer = FancyBboxPatch((2,1), 6, 3,
                       boxstyle="round,pad=0.4",
                       edgecolor='#cfa97e',
                       facecolor='#f4d9b6')
ax.add_patch(outer)

# -------------------------
# Inner boxes (top row)
# -------------------------
draw_box(2.5, 3, 1.5, 0.8, "Cleaning")
draw_box(4.25, 3, 1.5, 0.8, "Splitting")
draw_box(6, 3, 1.5, 0.8, "Vocabulary\nLookup")

# -------------------------
# Inner boxes (bottom row)
# -------------------------
draw_box(2.5, 1.8, 1.5, 0.8, "Add Special\nTokens")
draw_box(4.25, 1.8, 1.5, 0.8, "Generate\nAttention Masks")
draw_box(6, 1.8, 1.5, 0.8, "Generate Token\nType IDs")

# -------------------------
# Input & Output
# -------------------------
draw_round_box(0.5, 2.2, 1.3, 0.8, "Input Sentence", fc='#e91e63')
draw_round_box(8.5, 2.2, 1.8, 0.8, "Final Encoding\nfor Model", fc='#e91e63')

# -------------------------
# Arrows
# -------------------------
arrow(1.8, 2.6, 2, 2.6)
arrow(8, 2.6, 8.5, 2.6)

# -------------------------
# Final styling
# -------------------------
ax.set_xlim(0, 11)
ax.set_ylim(1, 6)
ax.axis('off')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Categories (same as image)
labels = [
    "Semantic understanding",
    "Fluency",
    "Sensitivity to word order",
    "Synonym handling",
    "Alignment with human judgment",
    "Computational cost"
]

# Values (approximated from your image)
rouge_bleu = [1, 1, 4, 1, 2, 1]
meteor      = [2, 1, 3, 4, 3, 2]
bertscore   = [5, 3, 1, 4, 5, 4]
bartscore   = [5, 5, 1, 5, 5, 5]

# Convert to circular
angles = np.linspace(0, 2*np.pi, len(labels), endpoint=False)

# Close the loop
def close(vals):
    return np.concatenate([vals, [vals[0]]])

angles = np.concatenate([angles, [angles[0]]])

# Plot
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, polar=True)

# Plot each metric
ax.plot(angles, close(rouge_bleu), label="ROUGE & BLEU")
ax.fill(angles, close(rouge_bleu), alpha=0.1)

ax.plot(angles, close(meteor), label="METEOR")
ax.fill(angles, close(meteor), alpha=0.1)

ax.plot(angles, close(bertscore), label="BERTScore")
ax.fill(angles, close(bertscore), alpha=0.1)

ax.plot(angles, close(bartscore), label="BARTScore")
ax.fill(angles, close(bartscore), alpha=0.1)

# Labels
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels, fontsize=9)

# Scale (0–5 like image)
ax.set_yticks([1,2,3,4,5])
ax.set_ylim(0,5)

# Title & legend
plt.title("Comparison of NLP Evaluation Metrics (Scale: 0–5)", pad=20)
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))

plt.show()
```
